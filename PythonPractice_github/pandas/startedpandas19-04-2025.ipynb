```python
import matplotlib.pyplot as plt

df['Order Date'] = pd.to_datetime(df['Order Date'])


df.set_index('Order Date')['Total Revenue'].plot()
plt.title('Total Revenue Over Time')
plt.xlabel('Date')
plt.ylabel('Total Revenue')
plt.show()
```

```python
df.set_index('Order Date')['Total Revenue'].plot(title='Revenue Over Time', figsize=(12, 6))
plt.xlabel('Date')
plt.ylabel('Revenue')
plt.grid(True)
plt.show()
```

```python
df.groupby('Region')['Total Profit'].sum().plot(kind='bar', title='Total Profit by Region')
plt.ylabel('Profit')
plt.tight_layout()
plt.show()
```

```python
df['Units Sold'].plot(kind='hist', bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Units Sold')
plt.xlabel('Units Sold')
plt.show()
```

```python
df['Sales Channel'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title('Sales Channel Breakdown')
plt.ylabel('')
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.plot(x, y)             
plt.title("Simple Line Plot")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.grid(True)
plt.show()
```

```python
x = ['A', 'B', 'C']
y = [10, 15, 7]

plt.bar(x, y, color='orange')
plt.title("Bar Chart Example")
plt.show()
```

```python
data = [1, 1, 2, 3, 3, 3, 4, 5, 6, 6, 7, 8, 9, 10]

plt.hist(data, bins=5, color='skyblue', edgecolor='black')
plt.title("Histogram")
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 6, 5]

plt.scatter(x, y, color='green')
plt.title("Scatter Plot")
plt.show()
```

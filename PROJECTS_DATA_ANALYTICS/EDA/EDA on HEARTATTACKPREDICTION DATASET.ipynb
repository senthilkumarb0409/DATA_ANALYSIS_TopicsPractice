```python
import numpy as np

# Linear regression: fit a line (y = m*x + b)
Slope, Intercept = np.polyfit(heart_df['Age'], heart_df['Blood Pressure'], 1)

print("Slope:", Slope)
print("Intercept:", Intercept)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))

# Scatter plot of actual data
plt.plot(heart_df['Age'], heart_df['Blood Pressure'], 'o', label='Data')

# Plotting the linear regression line
plt.plot(heart_df['Age'], Slope * heart_df['Age'] + Intercept, color='red', label='Fit Line')

# Axis labels
plt.xlabel("Age", fontsize=15)
plt.ylabel("Blood Pressure", fontsize=15)

# Optional: add legend
plt.legend()

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Fit linear regression: Blood Pressure ~ Cholesterol
Slope1, Intercept1 = np.polyfit(heart_df['Cholesterol'], heart_df['Blood Pressure'], 1)

print("Slope:", Slope1)
print("Intercept:", Intercept1)

# Plot data points
plt.figure(figsize=(8, 5))
plt.plot(heart_df['Cholesterol'], heart_df['Blood Pressure'], 'o', label='Data')

# Plot regression line
plt.plot(heart_df['Cholesterol'], Slope1 * heart_df['Cholesterol'] + Intercept1, color='red', label='Fit Line')

plt.xlabel("Cholesterol", fontsize=15)
plt.ylabel("Blood Pressure", fontsize=15)
plt.title("Linear Fit: Cholesterol vs Blood Pressure")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Fit linear regression: Blood Pressure ~ Age
Slope2, Intercept2 = np.polyfit(heart_df['Age'], heart_df['Blood Pressure'], 1)  # 1 denotes linear fit (Age^1)

print("Slope:", Slope2)
print("Intercept:", Intercept2)

# Plot data points
plt.figure(figsize=(8, 5))
plt.plot(heart_df['Age'], heart_df['Blood Pressure'], 'o', label='Data')

# Plot regression line
plt.plot(heart_df['Age'], Slope2 * heart_df['Age'] + Intercept2, color='red', label='Fit Line')

plt.xlabel("Age", fontsize=15)
plt.ylabel("Blood Pressure", fontsize=15)
plt.title("Linear Fit: Age vs Blood Pressure")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Linear regression: Max Heart Rate ~ Age
Slope3, Intercept3 = np.polyfit(heart_df['Age'], heart_df['Max Heart Rate'], 1)  # 1 = linear fit

print("Slope:", Slope3)
print("Intercept:", Intercept3)

# Plot data points
plt.figure(figsize=(8, 5))
plt.plot(heart_df['Age'], heart_df['Max Heart Rate'], 'o', label='Data')

# Plot regression line using Slope3 and Intercept3
plt.plot(heart_df['Age'], Slope3 * heart_df['Age'] + Intercept3, color='red', label='Fit Line')

plt.xlabel("Age", fontsize=15)
plt.ylabel("Max Heart Rate", fontsize=15)
plt.title("Linear Fit: Age vs Max Heart Rate")
plt.legend()
plt.show()
```
